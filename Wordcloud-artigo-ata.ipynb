## Código para retirar as stopwords em português.

```python
# Verificando as stopwords
sport = set(nltk.corpus.stopwords.words('portuguese'))
```

```python
sport
```

```python
# Adicionando uma lista de stopwords
sport2 = nltk.corpus.stopwords.words('portuguese')
novastop = ['Assim','Nesse', 'leite', '\n', ' ']
sport2.extend(novastop)
```

```python
# Gerando nuvem de palvras sem as stop words

wc3=WordCloud(stopwords=sport2)
wc3.generate(artigo2)
plt.figure(figsize=(10,10))
plt.imshow(wc3)
```

```python
# Mundando o plano de fundo
wc3=WordCloud(background_color='white', stopwords=sport2)
wc3.generate(artigo2)
plt.figure(figsize=(10,10))
plt.imshow(wc3)
```

```python
# gerando uma imagem da nuvme de palavras
wc3.to_file('nuvem_artigo_ata.png')
```
